```python
import json
```

```python
data = json.load(open("src/assets/data/foodData.bin.mid.json"))
```

```python
data
```

```python
profit_data = []
for food in data:
    revenue = food['PricePerSeving'] * food['Servings']
    cost = food['Price']
    profit = revenue - cost
    
    profit_data.append({
        'name': food['Name'],
        'profit': profit,
        'cost': cost,
        'revenue': revenue,
        'servings': food['Servings'],
        'price_per_serving': food['PricePerSeving'],
        'level': food['UnlockLevel']
    })
```

```python
profit_data
```

```python
profit_data.sort(key=lambda x: x['profit'], reverse=True)
    
# Print top 20 most profitable foods
print("Top 20 Most Profitable Foods:")
print("=" * 80)
print(f"{'Name':<30} | {'Profit':<10} | {'Revenue':<10} | {'Cost':<8} | {'Level':<5}")
print("-" * 80)

for food in profit_data[:20]:
    print(f"{food['name']:<30} | {food['profit']:<10} | {food['revenue']:<10} | {food['cost']:<8} | {food['level']:<5}")
```
